```python
import pandas as pd

df_odre=pd.read_csv('api_odre.csv')
df_ecoenergie=pd.read_csv('web_scraping_ecoE.csv')
df_energy_sources=df_odre.merge(df_ecoenergie,how='left',on='Source')

# Calculates the weighted average for emissions
total_powerage=0
total_emissions=0
for s in df_energy_sources.values:
  total_powerage+=float(s[1])
  total_emissions+=float(s[1])*float(s[3])
wavg_emissions=total_emissions/total_powerage
print(f"Weighted average : {wavg_emissions} gCO2e/KWh")
df_energy_sources
```
